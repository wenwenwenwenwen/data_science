## Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("classic")
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g') # (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # 0~1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # 16
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.5)) # RGB tuple 0~1
plt.plot(x, np.sin(x - 5), color='chartreuse') # HTML
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

```python
plt.plot(x, x + 0, linestyle='-')
plt.plot(x, x + 1, linestyle='--')
plt.plot(x, x + 2, linestyle='-.')
plt.plot(x, x + 3, linestyle=':')
```

```python
# combine color and linestyle
plt.plot(x, x + 0, '-g') # green
plt.plot(x, x + 1, '--c') # cyan
plt.plot(x, x + 2, '-.k') # black
plt.plot(x, x + 3, ':r') # red
```

```python
plt.plot(x, np.sin(x))
# set axes limit
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
# reverse y axes
plt.ylim(1.2, -1.2)
```

```python
plt.plot(x, np.sin(x))
# [xmin, xmax, ymin, ymax]
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
# set axis tight
plt.axis("tight")
```

```python
plt.plot(x, np.sin(x))
# set x axis equal to y axis
plt.axis("equal")
```
