## Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("classic")
```

```python
x = np.linspace(0, 10, 100)
```

#### Set Tag

```python
plt.plot(x, np.sin(x))
plt.title("A Sin Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), "-g", label="sin(x)")
plt.plot(x, np.cos(x), ":b", label="cos(x)")
plt.axis("equal")
# add legend
plt.legend()
```

```python
# plt.xlabel -> ax.set_xlabel, etc.

# set all attribute at once
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel="x", ylabel="sin(x)",
       title="A Simple Plot")
```
